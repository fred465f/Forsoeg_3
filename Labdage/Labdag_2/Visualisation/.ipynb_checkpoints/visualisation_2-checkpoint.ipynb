```python
# Standard Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Plot settings
plt.rc("axes", labelsize=18)
plt.rc("xtick", labelsize=16, top=True, direction="in")
plt.rc("ytick", labelsize=16, right=True, direction="in")
plt.rc("axes", titlesize=18)
plt.rc("legend", fontsize=16)
plt.rc("figure", figsize=(7, 5), titlesize=18)
```

```python
# Functions
def make_ax(data, x, y, xlabel = 'x', ylabel = 'y', title = 'Title', figsize = (10, 10), label = None):
    # Make fig and ax
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    
    # Plot data
    ax.scatter(data[x], data[y], label = label, color = 'red')

    # Set title and axis labels
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    # Return fig and ax
    return fig, ax

def plot_ax(ax, legend = False, grid = False, save = False, show = False, close = False, filename = 'default.pdf'):
    if grid:
        ax.grid()
    if legend:
        ax.legend()
    if save:
        plt.savefig(filename)
    if show:
        plt.show()
    if close:
        plt.close()
```

```python
# Load data
data = pd.read_csv('../Cleaned_Data/Final_data_2/eks_2_cleaned_data.csv')
```

```python
# Make ax and fig
fig, ax = make_ax(data, x = 'Angle', y = 'Mean Peak Value', figsize = (12, 8), 
                      title = 'Angle vs. Max Voltage', 
                      xlabel = r'Angle ($^{\circ}$)', ylabel = 'Max Voltage (V)')

# Plot ax
plot_ax(ax, grid = True, save = True, filename = 'angle_vs_max_voltage.pdf')
```
