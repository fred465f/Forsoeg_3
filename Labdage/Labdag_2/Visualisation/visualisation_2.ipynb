```python
# Standard Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

```python
# Plot settings
plt.rc("axes", labelsize=18)
plt.rc("xtick", labelsize=16, top=True, direction="in")
plt.rc("ytick", labelsize=16, right=True, direction="in")
plt.rc("axes", titlesize=18)
plt.rc("legend", fontsize=16)
plt.rc("figure", figsize=(7, 5), titlesize=18)
```

```python
# Functions
def make_ax(data, x, y, xlabel = 'x', ylabel = 'y', xerr = 0, title = 'Title', figsize = (10, 10), label = None):
    # Make fig and ax
    fig, axs = plt.subplots(2, 1, figsize=figsize, sharex=True)
    
    # Plot data
    axs[0].errorbar(data[x], data[y], xerr = xerr, label = label, color = '#00B3AD', fmt="o", capsize = 2)

    # Set title and axis labels
    axs[0].set_title(title)
    axs[1].set_xlabel(xlabel)
    axs[0].set_ylabel(ylabel)
    axs[1].set_ylabel('Voltage (V)')
    
    # Remove horizontal space between axes
    fig.subplots_adjust(hspace=0)
    
    # Return fig and ax
    return fig, axs

def plot_ax(ax, legend = False, grid = False, save = False, show = False, close = False, filename = 'default.pdf'):
    for i in range(2):
        if grid:
            axs[i].grid()
        if legend:
            axs[i].legend()
    if save:
        plt.savefig(filename)
    if show:
        plt.show()
    if close:
        plt.close()
        
def abs_cos(x, a, b, c, d):
    return a * abs(np.cos(x * b + c)) + d
```

```python
# Load data
data = pd.read_csv('../Cleaned_Data/Final_data_2/eks_2_cleaned_data.csv')

# Variables
angle_error = np.pi/180 #degree
```

```python
# Make ax and fig
fig, axs = make_ax(data, x = 'Angle', y = 'Mean Peak Value', xerr = angle_error, figsize = (12, 10), 
                      title = 'Angle vs. Max Voltage', label = 'Data', 
                      xlabel = r'Angle (Radians)', ylabel = 'Max Voltage (V)')

# Make fit
popt, pcov = curve_fit(abs_cos, data['Angle'], data['Mean Peak Value'],
                      p0 = [4, 1/70, 1/2, 6.5])

# Plot fit
abs_coss = [abs_cos(x, popt[0], popt[1], popt[2], popt[3]) for x in data['Angle']]
axs[0].plot(data['Angle'], abs_coss, color = 'black', 
        linestyle = 'dashed', label = r'Fit for: $a \cdot \|\cos(b \cdot x + c)\| + d$')

# Plot residuals
residuals = [x - y for x, y in zip(list(data['Mean Peak Value']), abs_coss)]
axs[1].errorbar(data['Angle'], residuals, xerr = angle_error, color = "#006663", 
                fmt="o", capsize = 2, label = 'Residuals')

# Plot ax
plot_ax(axs, grid = True, save = True, legend = True, filename = '../../../Final_graphs/angle_vs_max_voltage.pdf')
```

```python
errors = np.sqrt(np.diag(pcov))
for i, x in enumerate(['a', 'b', 'c', 'd']):
    print(f'Value of {x} was found to be: {popt[i]} +- {errors[i]} \n')
```
